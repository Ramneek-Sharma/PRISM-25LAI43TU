```python
# baseline model prompt eng

import json
import torch
import pandas as pd
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from huggingface_hub import notebook_login


notebook_login()

models_to_test = [
    "google/gemma-2b-it",
    "Qwen/Qwen2.5-1.5B-Instruct",
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit"
]

with open("dataset.json", "r") as f:
    dataset = json.load(f)

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16
)

system_prompt = """
You are a Smart Home Controller.

Convert instructions into:
[Device: ...] | [Action: ...]

Rules:
- Do not add extra fields
- Use IGNORE for negation
- Use EXCLUDE for exceptions
- Output ONLY one line
"""


def parse_output(output):
    try:
        parts = output.split("|")
        device = parts[0].replace("[", "").replace("]", "").strip()

        action = None
        for p in parts:
            if "Action" in p:
                action = p.split(":")[1].replace("]", "").strip()

        return device, action
    except:
        return None, None


def run_all_baselines():
    all_results = []

    for model_name in models_to_test:
        print(f"\n🚀 Baseline for: {model_name}")

        tokenizer = AutoTokenizer.from_pretrained(model_name)
        tokenizer.pad_token = tokenizer.eos_token

        model = AutoModelForCausalLM.from_pretrained(
            model_name,
            quantization_config=quant_config,
            device_map="auto"
        )


        for category in dataset["categories"]:
            for example in category["examples"][:2]:

                prompt = f"{system_prompt}\nUser: {example['instruction']}\nAssistant:"
                inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

                with torch.no_grad():
                    outputs = model.generate(
                        **inputs,
                        max_new_tokens=40,
                        do_sample=False
                    )

                prediction = tokenizer.decode(
                    outputs[0][inputs.input_ids.shape[1]:],
                    skip_special_tokens=True
                ).strip()


                pred_device, pred_action = parse_output(prediction)
                true_device, true_action = parse_output(example["ideal_output"])

                device_match = int(pred_device == true_device)
                action_match = int(pred_action == true_action)
                exact_match = int(prediction == example["ideal_output"])

                all_results.append({
                    "Model": model_name,
                    "Category": category["category"],
                    "Instruction": example["instruction"],
                    "Ideal": example["ideal_output"],
                    "Prediction": prediction,
                    "Device_Match": device_match,
                    "Action_Match": action_match,
                    "Exact_Match": exact_match
                })


        del model
        torch.cuda.empty_cache()

    return all_results



results = run_all_baselines()
pd.DataFrame(results).to_csv("triple_baseline_results.csv", index=False)

print("✅ Done!")
```

```python
from unsloth import FastLanguageModel
import torch
from trl import SFTTrainer
from transformers import TrainingArguments
from datasets import Dataset
import json

# 1. Load model
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/gemma-2b-it-bnb-4bit",
    max_seq_length=2048,
    load_in_4bit=True,
)

# 2. Apply LoRA
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
)

# 3. Load dataset
with open("dataset.json", "r") as f:
    raw_data = json.load(f)

formatted_data = []

for cat in raw_data["categories"]:
    for ex in cat["examples"]:
        text = f"""### Instruction:
{ex['instruction']}

### Response:
{ex['ideal_output']}"""
        formatted_data.append({"text": text})

dataset = Dataset.from_list(formatted_data)

print("Dataset size:", len(dataset))

# 4. Trainer
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    max_seq_length=2048,
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        max_steps=100,
        learning_rate=2e-4,
        logging_steps=5,
        output_dir="outputs",
        fp16=not torch.cuda.is_bf16_supported(),
        bf16=torch.cuda.is_bf16_supported(),
    ),
)

# 5. Train
trainer.train()
```

```python
from unsloth import FastLanguageModel
import torch
import json
import pandas as pd



FastLanguageModel.for_inference(model)


with open("dataset.json", "r") as f:
    dataset = json.load(f)

def parse_output(text):
    try:
        parts = text.split("|")
        device = parts[0].strip()
        action = parts[1].strip() if len(parts) > 1 else ""
        return device, action
    except:
        return "ERROR", "ERROR"

results = []


for category in dataset["categories"]:
    for example in category["examples"]:

        prompt = f"""### Instruction:
{example['instruction']}

### Response:
"""

        inputs = tokenizer(
            prompt,
            return_tensors="pt",
            padding=True,
            truncation=True
        )

        inputs = {k: v.to(model.device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=50,
                do_sample=False,
                use_cache=False
            )

        prediction = tokenizer.decode(
            outputs[0][inputs["input_ids"].shape[1]:],
            skip_special_tokens=True
        ).strip()


        prediction = prediction.split("###")[0].strip()
        prediction = prediction.split("\n")[0].strip()


        pred_device, pred_action = parse_output(prediction)
        true_device, true_action = parse_output(example["ideal_output"])

        results.append({
            "Model": "Gemma",
            "Category": category["category"],
            "Device_Match": int(pred_device == true_device),
            "Action_Match": int(pred_action == true_action),
            "Exact_Match": int(prediction == example["ideal_output"])
        })


df = pd.DataFrame(results)

print("\n📊 GEMMA RESULTS:")
print(df[["Device_Match", "Action_Match", "Exact_Match"]].mean())

print("\n📊 CATEGORY-WISE:")
print(df.groupby("Category")[["Device_Match", "Action_Match", "Exact_Match"]].mean())
```
